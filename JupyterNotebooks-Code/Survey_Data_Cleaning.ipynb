```python
# importing libraries
import pandas as pd
```

```python
#Load the dataset
df = pd.read_csv('Mental_Health_Survey.csv')
df.head()
```

```python
#Merging Constantly and Frequently to Frequently
col_1 = 'How often do you feel stressed about your academic responsibilities?'
old_value1 = 'Constantly'
new_value1 = 'Frequently'

#Replacing the old value with the new value in the specified column
df[col_1] = df[col_1].replace(old_value1, new_value1)
df.head()
```

```python
#Merging Constantly and Frequently to Frequently AND Sometimes and Occasionally to Occasionally
col_2 = 'How often do you feel distracted or unfocused while studying due to symptoms mentioned in question above?'
old_value2 = 'Sometimes'
new_value2 = 'Occasionally'

col_3 = 'How often do you feel distracted or unfocused while studying due to symptoms mentioned in question above?'
old_value3 = 'Constantly'
new_value3 = 'Frequently'

#Replacing the old value with the new value in the specified column
df[col_2] = df[col_2].replace(old_value2, new_value2)

df[col_3] = df[col_3].replace(old_value3, new_value3)
df.head()
```

```python
#Merging Maybe and Yes to Yes
col_4 = 'Do you think there’s a negative perception around seeking assistance or talking about mental health with your peers?'
old_value4 = 'Maybe'
new_value4 = 'Yes'

#Replacing the old value with the new value in the specified column
df[col_4] = df[col_4].replace(old_value4, new_value4)
df.head()
```

```python
#Saving file to CSV
df.to_csv('Mental_Health_Survey_Cleaned.csv')
```
